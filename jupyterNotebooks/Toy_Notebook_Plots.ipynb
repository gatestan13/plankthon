```python
import matplotlib.pyplot as plt
import numpy as np
import matlab as ml
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
fig, (ax1, ax5) = plt.subplots(nrows=1, ncols=2, figsize=(8, 8), subplot_kw=dict(polar=True))
ax1.plot(x, y)
ax1.axis('off')
ax5.plot(y, x)
ax5.axis('off')
```
